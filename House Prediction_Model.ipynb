```python
# Here we are dropping the lib.....
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#here we reading the data
df = pd.read_excel('1553768847_housing.xlsx')
df
```

```python
# lookinf towards all the info...releated to data
df.info()
```

```python
treating nan values .......
df.isnull().sum()
```

```python
# we will here drop null values and will save here the data by using the inplace funtion
df.dropna(inplace=True)
```

```python
#you can see that the values are dropped here
df.info()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['median_house_value'], axis = 1)
y = df['median_house_value']

X_train,X_test,y_train,y_test = train_test_split(X,y , test_size = 0.20)
```

```python
#here we have join the X and Y
Train_data = X_train.join(y_train)
```

```python
Train_data
```

```python
#looking if there is any relations
Train_data.hist(figsize = (15,8))
```

```python
#perperation of heatmap for looking into the correleations between the variables
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.heatmap(Train_data.corr(), annot=True, cmap='YlGnBu')
plt.show()
```

```python
# Conversion to logarthmic values
Train_data['total_rooms'] = np.log(Train_data['total_rooms'] +1 )
Train_data['trotal_bedrooms'] = np.log(Train_data['total_bedrooms'] +1 )
Train_data['population'] = np.log(Train_data['population'] +1 )
Train_data['households'] = np.log(Train_data['households'] +1 )
```

```python
Train_data.hist(figsize = (15,8))
```

```python
#We can use rhe ocean proximity data and can add it as a new varible
Train_data = Train_data.join(pd.get_dummies(Train_data.ocean_proximity)).drop(['ocean_proximity'], axis = 1)
```

```python
Train_data
```

```python
#perperation of heatmpa for looking again the correleations between the variables
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.heatmap(Train_data.corr(), annot=True, cmap='YlGnBu')
plt.show()
```

```python
# Bleow the graph is showing that most expensive houses are near the oceans
plt.figure(figsize=(15, 8))
sns.scatterplot(x='latitude', y='longitude', data=Train_data, hue='median_house_value', palette='coolwarm')
plt.show()
```

```python
#Feature Engg.
Train_data['bedroom_ratios']  = Train_data['total_bedrooms'] / Train_data['total_rooms']
Train_data['household_rooms'] = Train_data['total_rooms'] / Train_data['households']
```

```python
#perperation of heatmpa for looking again the correleations between the variables
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.heatmap(Train_data.corr(), annot=True, cmap='YlGnBu')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = Train_data.drop(['median_house_value'], axis = 1)
y = Train_data['median_house_value']

X_train,X_test,y_train,y_test = train_test_split(X,y , test_size = 0.20)

lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
Test_data = X_train.join(y_train)

Test_data['total_rooms'] = np.log(Test_data['total_rooms'] +1 )
Test_data['trotal_bedrooms'] = np.log(Test_data['total_bedrooms'] +1 )
Test_data['population'] = np.log(Test_data['population'] +1 )
Test_data['households'] = np.log(Test_data['households'] +1 )

Test_data['bedroom_ratios']  = Test_data['total_bedrooms'] / Test_data['total_rooms']
Test_data['household_rooms'] = Test_data['total_rooms'] / Test_data['households']
```

```python
Test_data
```

```python
lm.score(X_test, y_test)
```
